```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyvista as pv
import scipy.io as sio
```

```python
perm = sio.loadmat('simulations3D/data_1272_64x64x8.mat', simplify_cells=True)['perm']
print('Perm: {}'.format(perm.shape))
```

```python
mult = 620
fig, axs = plt.subplots(3, 8, figsize=(20,6), sharex=True, sharey=True)
for i in range(3):
    for j in range(8):
        ax = axs[i,j]
        k = i*mult
        ax.imshow(perm[k,...,j], cmap='jet')
        ax.set(xticks=[], yticks=[])
        ax.set(ylabel='R {}'.format(k)) if j==0 else None
plt.tight_layout()
plt.show()
```

```python
p = pv.Plotter()
p.add_mesh(np.flip(perm[1240]), cmap='jet')
p.show(jupyter_backend='static')
```
